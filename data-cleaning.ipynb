```python
import pandas as pd
```

```python
df = pd.read_csv('Highest-career-batting-average-in-Tests.csv')
```

```python
df
```

### rename multiple columns in a list

```python
df.rename(columns={'Mat':'Matches','NO': 'Not_Outs', 'HS': 'Highest_Inns_Score', 'BF': 'Balls_faced', 'SR':'Batting_Strike_Rate'}, inplace=True)
```

```python
df.head()
```

### check null value

```python
df.isnull().any()
```

```python
df[df['Balls_faced'].isna()==1]
```

```python
df.fillna(0, inplace=True)
```

### drop duplicate

```python
df.duplicated().any()
```

```python
df[df['Player'].duplicated()==1]
```

```python
df[df['Player'].isin(['RG Pollock (SA)', 'ED Weekes (WI)', 'JH Kallis (ICC/SA)'])]
```

```python
df.drop_duplicates(inplace=True)
df[df['Player'].duplicated()==1]
```

### split up Span into Start and End Date

```python
df['Span'].str.split(pat='-')
```

```python
df['Rookie_Year'] = df['Span'].str.split(pat='-').str[0]
df['Final_Year'] = df['Span'].str.split(pat='-').str[1]
```

```python
df.head()
```

### drop span columns

```python
df.drop(['Span'], axis=1, inplace=True) 
df.head()
```

### split the country from player

```python
df['Player'].str.split(pat='(')
```

```python
df['Country'] = df['Player'].str.split(pat='(').str[1]
```

```python
df['Country']
```

```python
df['Country'] = df['Country'].str.split(pat=')').str[0]
```

```python
df['Country']
```

```python
df['Player'] = df['Player'].str.split(pat='(').str[0]
df.head()
```

### change data type

```python
df.dtypes
```

```python
df['Highest_Inns_Score'] = df['Highest_Inns_Score'].str.split(pat='*').str[0]
df['Highest_Inns_Score']
```

```python
df['Highest_Inns_Score'] = df['Highest_Inns_Score'].astype('int')
```

```python
df.dtypes
```

```python
df = df.astype({'Rookie_Year':'int', 'Final_Year':'int'})
```

```python
df.dtypes
```

```python
df.head()
```

```python
df['Balls_faced'] = df['Balls_faced'].str.split(pat='+').str[0]
```

```python
df['Balls_faced'] = df['Balls_faced'].astype('int')
```

```python
df.isnull().any()
```

```python
df['Balls_faced'] = df['Balls_faced'].fillna(0)
```

```python
df['Balls_faced'] = df['Balls_faced'].astype('int')
```

```python
df.dtypes
```

### Build out Career_Length Column and hint convert rookie/final year and subtract

```python
df['Career_Length'] = df['Final_Year'] - df['Rookie_Year']
df.head()
```

```python
# Question 1 -> Cricketers in this DF what is the Average Career Length
```

```python
df['Career_Length'].mean()
```

```python
# Question 2 -> AVG Batting Strike Rate for cricketers who played over 10 years
```

```python
df[df['Career_Length'] > 10]['Batting_Strike_Rate'].mean()
```

```python
# Question 3 -> find number of cricketers who played before 1960
```

```python
df[df['Rookie_Year'] < 1960]['Player'].count()
```

```python
#Question 4 -> Max Highest Inns Score by country
```

```python
df.groupby('Country')['Highest_Inns_Score'].max().to_frame('Highest_Inns_Country').reset_index().sort_values('Highest_Inns_Country', ascending=False)
```
